```python
df_year = df[df['Year'] == 1968] 
print(df_year.head()) 
```

```python
# Use .loc[] to ensure you modify the dataframe correctly
df_year.loc[:, 'Value'] = pd.to_numeric(df_year['Value'], errors='coerce')

# Drop rows with NaN values in the 'Value' column
df_year = df_year.dropna(subset=['Value'])

# Check the cleaned data
print(df_year.head())
```

```python
import matplotlib.pyplot as plt

# Plotting the data for 1968
plt.figure(figsize=(12,6))
plt.bar(df_year['Country Name'], df_year['Value'], color='skyblue')
plt.xticks(rotation=90, ha='right')
plt.ylabel(f"Value in 1968")
plt.title(f"All Countries by Value in 1968")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# List of years you want to plot
years = [
    1968, 1969, 1970, 1971, 1972, 1973, 1974, 1975, 1976, 1977, 1978, 1979, 1980, 1981,
    1982, 1983, 1984, 1985, 1986, 1987, 1988, 1989, 1990, 1991, 1992, 1993, 1994, 1995,
    1996, 1997, 1998, 1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009,
    2010, 2011, 2012, 2013, 2014, 2015, 2016, 1960, 1961, 1962, 1963, 1964, 1965, 1966,
    1967
]

# Filter the dataframe for the required years and sum the 'Value' column per year
df_total_gdp = df[df['Year'].isin(years)]

# Aggregate the GDP per year (sum of 'Value' per year)
df_total_gdp_yearly = df_total_gdp.groupby('Year')['Value'].sum().reset_index()

# Plot the total GDP for each year
plt.figure(figsize=(14, 7))
plt.plot(df_total_gdp_yearly['Year'], df_total_gdp_yearly['Value'], marker='o', color='green')

# Customize the plot
plt.xticks(rotation=45)
plt.xlabel('Year')
plt.ylabel('Total GDP (Value)')
plt.title('Total GDP for Each Year')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter for South Africa and United States
df_south_africa = df[df['Country Name'] == 'South Africa']
df_usa = df[df['Country Name'] == 'United States']

# Plot the data for both countries
plt.figure(figsize=(12,6))

plt.plot(df_south_africa['Year'], df_south_africa['Value'], label='South Africa', color='blue', marker='o')
plt.plot(df_usa['Year'], df_usa['Value'], label='United States', color='green', marker='x')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Value (GDP or Indicator)')
plt.title('South Africa vs United States GDP over the Years')

# Show legend to distinguish between the countries
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```
